```python
def my_check_function (dataset):
    '''Функция для ознакомления и проверки даных'''
    print('Размер датасета (количество строк, столбцов):')
    print(dataset.shape)
    print('Как выглядят случайные 5 строк:')
    display(dataset.sample(5, random_state=42))
    print()
    print('Общая информация о данных, наименования столбцов, типы данных:')
    print(dataset.info())
    print()
    print('Есть ли пропуски, сколько их:')
    print(dataset.isna().sum())
    print('Доля пропущенных значений:')
    print(dataset.isna().mean())
    print()
    print('Есть ли явные дубликаты, сколько их:')
    print(dataset.duplicated().sum())
    print('Доля явных дубликатов:')
    print(dataset.duplicated().mean())
    print()
    print('Подробное описание данных:')
    display(dataset.describe())
```

```python
df['user_score'].value_counts()
```

Кроме чисел здесь имеется аббревиатура tbd: to be determined. Значит, в строках собозначением tbd мы имеем по сути пропуск: пользовательская оценка ещё не определена. 

Поскольку это значение помешает нам работать с данными в этом столбце (нам нужен числовой формат), в процессе преобразования удалим его:

```python
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
```

В некоторых столбцах у нас мало пропусков, а в некоторых очень много. Возможно, в первых столбцах (name, year_of_release, genre) пропуски имеют случайный характер, а в последних есть какая-то закономерность и они связаны между собой. Возьмём столбец с наибольшим количеством пропусков (user_score) и визуализируем, насколько пересекаются пропуски.

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
plt.style.use('seaborn-whitegrid')
plt.suptitle('Количество пропусков', fontweight='bold')

empty_df = pd.DataFrame(df.isna().sum()).reset_index().rename_axis(None, axis=1)
empty_df.columns=['columns', 'isna_sum']

ax1 = plt.subplot(1, 2, 1)
sns.barplot(data=empty_df, x='isna_sum', y='columns',  color='#e2c1b3')
plt.title('во всём датасете')
plt.xlabel('')
plt.ylabel('')

empty_user_score = pd.DataFrame(df.loc[df['user_score'].isna()].isna().sum()).reset_index().rename_axis(None, axis=1)
empty_user_score.columns=['columns', 'isna_sum']

ax2 = plt.subplot(1, 2, 2)
sns.barplot(data=empty_user_score, x='isna_sum', y='columns',  color='#e2c1b3')
plt.title('в строках с пропущенными пользовательскими оценками')
plt.xlabel('')
plt.ylabel('')

plt.tight_layout()
plt.show();
```

Как мы видим, пропуски сильно пересекаются: те игры, у которых нет пользовательских оценок, очень часто не имеют и оценок от критиков и часто не имеют рейтинга.

Попробуем разобраться с возможными причинами этой ситуации. 

**Гипотеза 1**: у игр нет оценок и рейтинга, потому что это не популярные игры. В этом случае мы увидим разницу в ***количестве продаж*** между играми с оценками и играми без оценок.

**Гипотеза 2**: оценок и рейтинга нет у игр с определённых ***платформ***. Если дело в этом, то мы увидим разницу в наименованиях платформ: у игр с оценками будут одни платформы, а у игр без оценок будут другие.

**Гипотеза 3**: дело во ***времени выхода*** игр. Недавно вышедшие игры ещё не успели получить оценок. В этом случае мы увидим, что игры без оценок имеют недавний год выпуска.

**Гипотеза 4**: ***жанр*** игры влияет на наличие оценки. В этом случае мы увидим, что у некоторых жанров есть заметная разница в количестве игр с оценками и без.

Добавим в датасет данные о суммарных продажах во всех регионах:

```python
df.head()
```

```python
df.insert(8, 'total_sales', df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales'])
```

Проверим результат:

```python
df.head()
```

Укажем в датасете признак: является ли строка "пустой" или "полной" по наличию пользовательской оценки.

```python
df['full_empty_user_score'] = df['user_score'].apply(lambda x: 'full' if x < 11 else 'empty')
```

Теперь проверим гипотезы. Для удобства визуализируем данные по всем гипотезам на одном графике из нескольких частей:

```python
plt.figure(figsize=(20, 15))
sns.set(font_scale=1.5)
plt.style.use('seaborn-whitegrid')
plt.suptitle('Отличия игр с пользовательскими оценками и без них', fontweight='bold')

sales = df[['full_empty_user_score','na_sales','eu_sales', 'jp_sales', 'other_sales', 'total_sales']]
sales = (sales.pivot_table(index='full_empty_user_score', values=sales.columns, aggfunc='sum')
            .T.reset_index().rename_axis(None, axis=1).sort_values(by='empty', ascending=False))
ax1 = plt.subplot(2, 2, 1)
sns.barplot(data=sales, x='full', y='index', ax=ax1, label='игры с пользовательскими оценками', color='#e2c1b3')
sns.lineplot(data=sales, x='empty', y='index', ax=ax1, label='игры без пользовательских оценок', color='#20597b')
plt.title('1. Количество продаж', fontweight='bold')
plt.xlabel('миллионы проданных копий')
plt.ylabel('')

platforms = df[['full_empty_user_score','platform', 'user_score']]
platforms = (platforms.pivot_table(index='platform', values=platforms.columns, aggfunc='count')
                    .reset_index().rename_axis(None, axis=1))
platforms.columns = ['platform', 'total_cnt', 'full_cnt']
platforms['empty_cnt'] = platforms['total_cnt'] - platforms['full_cnt']

ax2 = plt.subplot(2, 2, 2)
sns.barplot(data=platforms.sort_values(by='empty_cnt', ascending=False), y='platform', x='full_cnt', ax=ax2,
               label='игры с пользовательскими оценками', color='#e2c1b3')
sns.lineplot(data=platforms.sort_values(by='empty_cnt', ascending=False), y='platform', x='empty_cnt', ax=ax2, 
             label='игры без пользовательских оценок', color='#20597b')
plt.title('2. Платформы', fontweight='bold')
plt.xlabel('количество игр')
plt.ylabel('')

years =  df[['full_empty_user_score','year_of_release', 'user_score']]
years = (years.pivot_table(index='year_of_release', values=years.columns, aggfunc='count')
                    .reset_index().rename_axis(None, axis=1).sort_values(by='year_of_release'))
years.columns = ['year', 'total_cnt', 'full_cnt']
years['empty_cnt'] = years['total_cnt'] - years['full_cnt']
ax3 = plt.subplot(2, 2, 3)
sns.lineplot(data=years, x='year', y='full_cnt', ax=ax3, label='игры с пользовательскими оценками', color='#e2c1b3')
sns.lineplot(data=years, x='year', y='empty_cnt', ax=ax3, label='игры без пользовательских оценок', color='#20597b')
plt.title('3. Годы выхода игр', fontweight='bold')
plt.xlabel('')
plt.ylabel('количество игр')

genres = df[['full_empty_user_score','genre', 'user_score']]
genres = (genres.pivot_table(index='genre', values=genres.columns, aggfunc='count')
            .reset_index().rename_axis(None, axis=1))
genres.columns = ['genre', 'total_cnt', 'full_cnt']
genres['empty_cnt'] = genres['total_cnt'] - genres['full_cnt']
ax4 = plt.subplot(2, 2, 4)
sns.barplot(data=genres.sort_values(by='empty_cnt', ascending=False), y='genre', x='full_cnt', ax=ax4,
               label='игры с пользовательскими оценками', color='#e2c1b3')
sns.lineplot(data=genres.sort_values(by='empty_cnt', ascending=False), y='genre', x='empty_cnt', ax=ax4, 
             label='игры без пользовательских оценок', color='#20597b')
plt.title('4. Жанры', fontweight='bold')
plt.xlabel('количество игр')
plt.ylabel('')

plt.tight_layout()
plt.show();
```

Как мы видим, одной-единственной причины отсутствия пользовательских оценок нет, мы можем увидеть связь пропусков с разными факторами:

- ***игры без оценок в целом менее популярны*** и имеют меньшее количество миллионов проданных копий, чем игры с оценками. Здесь мы видим уверенную закономерность за исключением продаж в Японии. Возможно, пользователи из этого региона менее интегрированы в сообщество игроков всего мира и не всегда знают о сайтах, где можно выставить оценки.


- с платформами смешанная ситуация: у некоторых платформ больше игр с оценками, у других - больше игр без оценок. То есть явной закономерности нет, но некоторые ***платформы*** вносят свой вклад в наличие пропусков


- игр без оценок больше в ранние ***годы***: до 2000 года большая часть игр не имеет оценок, видимо, это связано с тем, что индустрия только развивалась в это время, и не было сайтов, где собирались бы такие оценки. Другой всплеск - 2007-2011 годов, возможно, связан с возросшей конкуренцией - игр уже стало много, и не все могут в этих условиях найти своё место под Солнцем (в душе пользователя).


- только ***жанры*** Action и Shooter имеют много игр с оценками. В остальном пользователи не особо спешат делиться своим мнением об играх других жанров.

Пропуски заполнять не будем, поскольку это может исказить достоверность наших данных.

### Актуальный период и "ящики с усами"

Посмотрим, как менялись продажи по платформам за последние годы и какие платформы в лидерах за последние годы:

**Вывод**:

- ***за последние годы по объёму продаж вперёд вышли PS4, PS3 и XOne***

Поскольку ранее мы выяснили, что в отрасли сменяются лидеры примерно раз в 5 лет, можно остановиться на этом периоде как актуальном и выделить данные за период с 2012 года. 

Но сами наши данные показывают, что в последние годы характер изменений стал другим, а лидеры последних лет начали свой рост с 2013 года. Поэтому актуальнее для прогноза выделить данные с 2013 года: будет и не много (чтобы не захватить отжившие тренды), и не мало (чтобы данных для анализа и прогнозов было достаточно).

## Проверка гипотез

Проверим пару гипотез.

**1. Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые**

Сформулируем нулевую гипотезу:
- средние пользовательские рейтинги платформ Xbox One и PC равны

Альтернативная гипотеза:
 - средние рейтинги не равны
 
 
Это двусторонняя альтернативная гипотеза.

Для проверки гипотез нам нужно отобрать данные без пропусков:

```python
df_hyp = df_actual
df_hyp.dropna(subset=['user_score'], inplace=True)
```

Посмотрим, сколько у нас строк:

```python
len(df_hyp['user_score'])
```

Посмотрим на размер выборок:

```python
x_one = df_hyp.query('platform == "XOne"')
print(f'Размер выборки для платформы Xbox One: {len(x_one)}')
pc = df_hyp.query('platform == "PC"')
print(f'Размер выборки для платформы PC: {len(pc)}')
```

Поскольку мы работаем с гипотезами о равенстве средних двух генеральных совокупностей, спользуем метод scipy.stats.ttest_ind, куда в качестве аргументов передадим массивы, содержащие выборки для двух платформ. Размеры выборок близки, так что параметр equal_var=False нам использовать не придётся, он автоматически будет equal_var=True.

```python
alpha = 0.05

results = st.ttest_ind(x_one['user_score'], pc['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**:
- не получилось отвергнуть нулевую гипотезу, значит, статистически значимых различий в пользовательских оценках этих платформ нет

**2. Гипотеза: средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные**

Сформулируем нулевую гипотезу:
- средние пользовательские рейтинги жанров Action и Sports равны

Альтернативная гипотеза:
 - средние рейтинги не равны
  
  
Это снова двусторонняя альтернативная гипотеза. Хотя в этот раз нас интересует, разные ли средние значения, нулевую гипотезу мы всё равно формулируем так, чтобы использовать равенство.

Посмотрим на размер выборок:

```python
action = df_hyp.query('genre == "Action"')
print(f'Размер выборки для жанра Action: {len(action)}')
sports = df_hyp.query('genre == "Sports"')
print(f'Размер выборки для жанра Sports: {len(sports)}')
```

Мы снова работаем с гипотезами о равенстве средних двух генеральных совокупностей, поэтому снова спользуем метод scipy.stats.ttest_ind, но для этой гипотезы используем параметр equal_var=False, потому что размеры выборок заметно отличаются.

```python
alpha = 0.05

results = st.ttest_ind(action['user_score'], sports['user_score'], equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**:
- в результате теста нулевая гипотеза отвергнута, значит, средние пользовательские рейтинги двух жанров скорее разные и при сборе бОльших данных тоже будут отличаться.

**Дополнительно визуализируем**, каковы наши имеющиеся данные, на которых мы проводили статистические тесты:

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
plt.style.use('seaborn-whitegrid')
plt.suptitle('Гипотеза 1: сравнение пользовательских оценок двух платформ', fontweight='bold')

ax1 = plt.subplot(2, 1, 1)
sns.boxplot(x=x_one['user_score'], color='#ceab85')
plt.title('Xbox One', fontweight='bold')
plt.xlabel('пользовательские оценки')

ax2 = plt.subplot(2, 1, 2, sharex=ax1)
sns.boxplot(x=pc['user_score'], color='#aab35c')
plt.title('PC', fontweight='bold')
plt.xlabel('пользовательские оценки')

plt.tight_layout()
plt.show();
```
