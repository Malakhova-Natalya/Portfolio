## Описание проекта

В данном исследовании проводится анализ текущей бизнес-ситуации небольшого магазина товаров для дома, проводится сегментация покупателей по профилю потребления и даются рекомендации бизнесу на основе проведённого анализа данных.

**Основные задачи**:
- проанализировать основные показатели бизнеса: суммы продаж, количество заказов, средний чек, сезонность
- провести сегментацию покупателей по профилям потребления
- дать развёрнутые рекомендации по каждому кластеру: когда и какие рекламные предложения рассылать, как увеличить количество продаж, сумму покупок, повысить возвращаемость покупателей

**План работы**:
- предобработка данных: изменение типов данных, проверка на скрытые дубликаты
- подготовка данных к исследованию: создание категорий товаров
- анализ показателей бизнеса: продажи по годам, сезонам, категориям товаров; среднее количество заказов на покупателя и средний чек
- сегментация покупателей по профилям потребления при помощи KMeans
- проверка качества разбивки кластеров при помощи статистических гипотез: различия в среднем чеке между кластерами при качественной разбивке должны быть статистическими значимыми
- описание "портрета" пользователя из каждого кластера, визуализация различий между кластерами
- рекомендации по предложениям и рассылкам для каждого кластера
- презентация
- дашборд

**Описание данных**:

В исследовании используются данные, предоставленные самим магазином товаров для дома. Это сведения о покупках за несколько лет: дата заказа, идентификатор покупателя, номер заказа, наименование продукта, количество, цена.

### Выделение года, месяца, времен года из дат

Для дальнейшего исследования нам пригодится информация о годе и месяце оформления заказов. Добавим такие данные в датасет:

```python
df['year'] = pd.to_datetime(df['date']).dt.year.astype(str)
df['month'] = pd.to_datetime(df['date']).dt.month
df['month_name'] = pd.to_datetime(df['date']).dt.month_name()
df.head()
```

Создадим функцию для определения названий времен года:

```python
def season (month_name):
    """
    Возвращает название времени года, исходя из названия месяца
    """
    winter_list = ['December', 'January', 'February']
    spring_list = ['March', 'April', 'May']
    summer_list = ['June', 'July', 'August']
    autumn_list = ['September', 'October', 'November']
    
    for name in winter_list:
        if name in month_name:
            return 'Winter'
    for name in spring_list:
        if name in month_name:
            return 'Spring'
    for name in summer_list:
        if name in month_name:
            return 'Summer'
    for name in autumn_list:
        if name in month_name:
            return 'Autumn'
```

Добавляем данные в датасет:

```python
df['season'] = df['month_name'].apply(season) 
```

И добавляем порядковый номер времени года (начиная с зимы):

```python
def season_number (season):
    """
    Возвращает порядковый номер времени года, исходя из названия 
    """

    if season =="Winter":
            return 1
    if season =="Spring":
            return 2
    if season =="Summer":
            return 3
    if season =="Autumn":
            return 4
```

```python
df['season_number'] = df['season'].apply(season_number) 
df.head()
```

Проведём некоторые расчёты. Узнаем, сколько покупателей оформило 1 заказ, 2 заказа и т.д.

```python
data = df.groupby('customer_id', as_index=False)['order_id'].nunique()
data = data.groupby('order_id', as_index=False).agg({'customer_id':'count'})
data.columns = ['orders_number', 'customers_cnt']
data
```

Напишем функцию, которая поможет нам категоризировать эти данные:

```python
def orders_number_category (orders_number):
    """
    Возвращает категорию количества заказов у одного пользователя,
    используя следующие правила:
    - '1 заказ', если у клиента за всю историю покупок 1 заказ;
    - '2 заказа', если 2 заказа;
    - 'от 3 до 5 заказов', если от 3 (включая 3) до 5 заказов (не включая 5);
    - 'более 5 заказов', если больше 5 заказов
    """
    if orders_number == 1:
        return '1 заказ'
    if orders_number == 2:
        return '2 заказа'
    if 3 <= orders_number < 5:
        return 'от 3 до 5 заказов'
    if orders_number >= 5:
        return 'более 5 заказов'
```

### Покупатели: средний чек

Рассчитаем, сколько всего денег потратил каждый покупатель:

```python
data = df.groupby('customer_id', as_index=False).agg({'order_receipt':'sum'}).sort_values(by='order_receipt')
data.head()
```

Напишем функцию для определения категорий потраченных сумм:

```python
def receipt_category (order_receipt):
    """
    Возвращает категорию общей потраченной суммы на все покупки каждого пользователя,
    используя следующие правила:
    - 'менее 500 руб.', если у клиента общая сумма покупок менее 500 руб.;
    - 'от 500 до 2000 руб.', если сумма от 500 (включая 500 руб.) до 2000 руб. (не включая 2000 руб.)
    -  и т.д.
    """
    if order_receipt < 500:
        return 'менее 500 руб.'
    if 500 <= order_receipt < 2000:
        return 'от 500 до 2000 руб.'
    if 2000 <= order_receipt < 10000:
        return 'от 2 до 10 тыс.руб.'
    if order_receipt >= 10000:
        return 'более 10 тыс.руб.'
```

Добавляем новый столбец:

```python
data['receipt_category'] = data['order_receipt'].apply(receipt_category) 
data.head()
```

Теперь напишем функцию, чтобы задать порядок полученных категорий от наименьших к наибольшим потраченным суммам:

```python
def number_category (receipt_category):
    """
    Возвращает номер категории общей потраченной суммы по возрастанию объёмов потраченных денег
    """
    if receipt_category == 'менее 500 руб.':
        return 1
    if receipt_category == 'от 500 до 2000 руб.':
        return 2
    if receipt_category == 'от 2 до 10 тыс.руб.':
        return 3
    if receipt_category == 'более 10 тыс.руб.':
        return 4
```

Применяем функцию:

```python
data['number_category'] = data['receipt_category'].apply(number_category) 
data.head()
```

## Сегментация покупателей по профилю потребления

### Подготовка данных 

Подготовим данные для кластеризации. Поскольку нам нужно сегментировать покупателей по уровню потребления, сгруппируем данные по customer_id и добавим данные о потреблении.

Посмотрим, данные о каком количестве покупателей у нас есть:

Для того, чтобы значения сильно не выбивались, создидим функцию, которая поможет нам брать в датасет о покупателях не сами значения, а категории значений (например 0 покупок/1 покупка/много покупок в какой-то ценовой категории):

```python
def category_number (column_name):
    """
    Возвращает категорию используя следующие правила:
    - 0, если в столбце значение 0,
    - 1, если значение 1, 
    - 2, если значений несколько
    """
    if column_name == 0:
        return 0
    elif column_name == 1:
        return 1
    elif column_name >= 2:
        return 2
```

Создаём датасет, где каждая строка - покупатель, а столбцы - это данные о его уровне потребления:

```python
df['receipt_category'] = df['order_receipt'].apply(receipt_category) 
df['receipt_category_number'] = df['receipt_category'].apply(number_category) 

customer = (df.groupby('customer_id', as_index=False)
                  .agg({'order_id':'nunique', 'receipt_category':'nunique',
                       'order_receipt':'mean'})
                 .sort_values(by='order_id'))     

order_receipt_1 = (df.query('receipt_category_number == 1').groupby('customer_id', as_index=False)
                   .agg({'order_id':'nunique'}))
order_receipt_2 = (df.query('receipt_category_number == 2').groupby('customer_id', as_index=False)
                   .agg({'order_id':'nunique'}))
order_receipt_3 = (df.query('receipt_category_number == 3').groupby('customer_id', as_index=False)
                   .agg({'order_id':'nunique'}))
order_receipt_4 = (df.query('receipt_category_number == 4').groupby('customer_id', as_index=False)
                   .agg({'order_id':'nunique'}))

data_frames = [customer, order_receipt_1, order_receipt_2, order_receipt_3, order_receipt_4]

customer_table = reduce(lambda  left,right: pd.merge(left,right,on=['customer_id'],
                                            how='outer'), data_frames).fillna(0)
customer_table.columns = ['customer_id', 'order_total', 'category_cnt', 'order_receipt_mean','order_receipt_cat_1', 
                          'order_receipt_cat_2', 'order_receipt_cat_3', 'order_receipt_cat_4']
customer_table_raw = customer_table
customer_table['receipt_mean_category'] = customer_table['order_receipt_mean'].apply(receipt_category)
customer_table['avg_check_category'] = customer_table['receipt_mean_category'].apply(number_category)
customer_table['category_1'] = customer_table['order_receipt_cat_1'].apply(category_number)
customer_table['category_2'] = customer_table['order_receipt_cat_2'].apply(category_number)
customer_table['category_3'] = customer_table['order_receipt_cat_3'].apply(category_number)
customer_table['category_4'] = customer_table['order_receipt_cat_4'].apply(category_number)
customer_table = customer_table.drop(['receipt_mean_category','order_receipt_cat_1','order_receipt_cat_2', 
                                      'order_receipt_cat_3', 'order_receipt_cat_4'], axis=1)
customer_table
```

### Статистические тесты

Нас интересует, сработают ли принципы разбивки клиентов на кластеры в дальнейшем, например, при увеличении числа клиентов. То есть, имеют ли кластеры разные уровни потребления случайно, или это статистически значимые различия.

Проверим статистические гипотезы:

- H0: статистически значимых различий в среднем чеке между двумя кластерами нет
- H1: различия есть

Нас интересует ситуация, когда нулевая гипотеза отвергается.

Мы сравним все кластеры парами между собой, и если везде нулевая гипотеза будет отвергнута, значит, мы нашли работающую закономерность разбиения на кластеры. В этом случае возможно будет дать "портрет" пользователя и эффективные рекомендации на основе анализа данных.

Выделим кластеры и вспомним, какого размера выборки:

```python
cluster_A = df.query('cluster =="A"')
cluster_B = df.query('cluster =="B"')
cluster_C = df.query('cluster =="C"')


print(f'Размер кластера А: {len(cluster_A)}')
print(f'Размер кластера B: {len(cluster_B)}')
print(f'Размер кластера C: {len(cluster_C)}')
```

Проверим на нормальность распределения выборки:

```python
alpha = st.norm.ppf(1-0.05)  # критический уровень статистической значимости

results = st.shapiro(cluster_A['order_receipt'])
p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
    
results = st.shapiro(cluster_B['order_receipt'])
p_value = results[1] 

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
    
results = st.shapiro(cluster_C['order_receipt'])
p_value = results[1] 

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально')
```

У нас в данных есть выбросы, поэтому применить scipy.stats.ttest_ind не сможем, проведем тест Манна-Уитни:

```python
alpha = st.norm.ppf(1-0.05)  # критический уровень статистической значимости

results = st.mannwhitneyu(cluster_A['order_receipt'], cluster_B['order_receipt'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

- сильнее всего медианный чек увеличивается в апреле и мае. Видимо, это связано с подготовкой к отпускам. Тогда же увеличивается количество заказов и следом сумма продаж.


- а вот август и сентябрь финансово неблагоприятные для магазина месяцы по всем показателям
 
 
- самая популярная ***категория товаров*** - это "цветы и растения". В этой категории почти 3000 заказов на сумму чуть более 3.5 млн.руб. Однако медианный чек не высокий: менее 1000 руб.


- также популярностью пользуются товары из категории "уборка": более 1000 заказов на общую сумму более 2 млн руб. К этой категории относится аномальный по сумме заказ.


- по медианному чеку категории "транспортировка" и "интерьер" занимают лидирующее положение: это более 1 тыс руб.


- самый низкий медианный чек в категориях "декор" и "хозяйство": это менее 500 руб.
