## Описание проекта

В данном проекте происходит оценка результатов A/B-теста. 

Было  внедрено улучшение рекомендательной системы, тест должен помочь оценить результаты изменений: помогает ли новинка увеличить активность пользователей и повысить продажи?

Группа А видит прежнюю версию, группа В - версию с улучшением. Ожидается, что  за 14 дней с момента регистрации в системе пользователи покажут улучшение конверсии на каждом шаге (от просмотра товаров до покупки) не менее, чем на 10%.
 
**Основные задачи**:
- оценить корректность проведения теста
- проанализировать результаты теста

**План работы**:
- открытие, знакомство с данными, предобработка
- проверка на пересечение тестовой аудитории с конкурирующим тестом
- проверка на совпадение теста и маркетинговых событий, другие проблемы временных границ теста
- исследовательский анализ, проверка стат.гипотез

**Описание данных**:

В нашем рапоряжении есть данные с действиями пользователей, техническое задание (ниже) и несколько вспомогательных датасетов.

### Аудитория теста

Удостоверимся (или нет), что нет пересечений с конкурирующим тестом и нет пользователей, участвующих в двух группах теста одновременно. 

Проверим равномерность распределения пользователей по тестовым группам.

```python
user_double_test = df_raw.groupby('user_id', as_index=False).agg({'ab_test':'nunique'})
user_double_test = user_double_test.query('ab_test > 1')
print(f'Количество пересечений с конкурирующим тестом: {len(user_double_test)}')
```

Посмотрим, в какие группы теста interface_eu_test попали наши "продублированные" пользователи:

```python
double_users = user_double_test['user_id']

participants_int_eu_test = participants.query('ab_test=="interface_eu_test"')
participants_int_eu_test.query('user_id in @double_users')['group'].value_counts()
```

Они попали в обе группы теста, значит, на некоторых участников нашего теста мог повлиять другой тест (что тоже снижает чистоту эксперимента).

```python
user_double_group = df.groupby('user_id', as_index=False).agg({'group':'nunique'})
user_double_group = user_double_group.query('group > 1')
print(f'Количество пользователей, участвующих в двух группах теста одновременно: {len(user_double_group)}')
```

```python
cnt_A = df.query('group == "A"')['user_id'].nunique()
cnt_B = df.query('group == "B"')['user_id'].nunique()
print(f'Количество пользователей в группе А: {cnt_A}')
print(f'Количество пользователей в группе B: {cnt_B}')
```

**Вывод**:

- у нас есть пересечения с конкурирующим тестом - 887 человек попали в два разных теста
- но никто не попал внутри одного теста в две разные группы одновременно
- количество пользователей в группах А и B неравномерно

Уберём из нашего датасета строки, в которых события превышают заданный тестом порог в 14 дней:

```python
df['days_between_event_and_registr'] = df['event_dt'] - df['first_date']
df['days_between_event_and_registr'] = df['days_between_event_and_registr'].astype(str)
df['days_between_event_and_registr'] = df['days_between_event_and_registr'].str.strip('days') 
df['days_between_event_and_registr'] = df['days_between_event_and_registr'].astype(int)
df = df.drop(df[df.days_between_event_and_registr > 14].index)

df['days_between_event_and_registr'].value_counts()
```

### Вывод

Оценка корректности выявила ряд проблем и несоответствий техническому заданию. К результатам теста стоит относиться с большой осторожностью.

### Динамика набора пользователей

Посмотрим, как соотносятся события с количеством новых пользователей. Если мы увидим скачок в группе А с 13 на 14 декабря, значит, увеличение числа событий связано с увеличением числа новых пользователей, зарегистрированных в этот период.

### Проверка статистической разницы долей z-критерием

Проведем тест на проверку гипотезы о статистической разнице долей.

**Н0**: конверсии по шагам в выборках равны

**Н1**: конверсии разные

Будем использовать функцию:

```python
alpha = 0.05/3
distr = st.norm(0, 1)  
def z_test(successes, trials): 
    p1 = successes[0]/trials[0]
    p2 = successes[1]/trials[1]
    
    print(successes[0],successes[1] , trials[0], trials[1])
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])
    difference = p1 - p2 
    
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1])) 
    
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    
    print('p-значение: ', p_value)
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между выборками есть значимая разница')
    else:
        print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать выборки разными')
```
