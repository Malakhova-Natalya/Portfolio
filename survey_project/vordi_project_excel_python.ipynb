## Сравнение различных категорий с помощью метрики "процент удовлетворения"

У нас есть ответы, оцифрованные в такой общей логике: 
- 1 - положительный ответ
- 0.5 - скорее положительный 
- 0 - затруднения в ответе/нейтрально/прочее
- -0.5 - скорее отрицательный
- -1 - отрицательный

Мы вычислили их сумму по каждому вопросу. То есть если бы у всех участников был положительный ответ (1), то сумма баллов была бы равна количеству участников. Например, 997 участников - и возможный максимум 997 баллов.

И возможно наоборот - если у всех участников проблемы в каком-то вопросе (-1), то максимально возможный минимум -997 баллов.

Скорее всего, в каждом вопросе мы будем видеть смешанную картину, но общая сумма может говорить о **преобладании** проблем или удовлетворения по каждому вопросу. Чем выше сумма - тем больше удовлетворения, чем меньше сумма - тем больше проблем.

Но чтобы понимать этот показатель **в контексте** (в разные годы количество участников разное), мы определим, какой процент составляет сумма баллов от количества участников. Так мы получим более объективный показатель - назовём его **% общей удовлетворённости**.

С его помощью мы сможем выявить вопросы и категории вопросов, где участники в большинстве своём испытывают проблемы.

Выделим в отдельный датафрейм те строки, где у нас есть оцифрованные данные и соответственно есть расчёты метрики "процент удовлетворения". Некоторые вопросы из блока о личных данных тоже имеют тип стандарт, но они нам сейчас не нужны, поэтому задаём два условия, чтобы отобрать нужные данные:

```python
df_21 = df_questions_21.query('tech_status == "стандарт" and perc_satisfaction_21 < 101') 
df_22 = df_questions_22.query('tech_status == "стандарт" and perc_satisfaction_22 < 101')
df_21
```

### Категории вопросов

Создадим датафрейм, сгруппировав данные по категориям вопросов:

```python
df_category_21 = (df_21.groupby('category_description', as_index=False)
                   .agg({'perc_satisfaction_21':'mean'}))
df_category_22 = (df_22.groupby('category_description', as_index=False)
                   .agg({'perc_satisfaction_22':'mean'}))
df_category = df_category_22.merge(df_category_21, on='category_description', how='left')
df_category['dynamic'] = df_category['perc_satisfaction_22'] - df_category['perc_satisfaction_21']
df_category = df_category.sort_values(by='perc_satisfaction_22', ascending=False)
```

Рассчитаем среднее:

```python
mean_2021 = df_category['perc_satisfaction_21'].mean()
mean_2022 = df_category['perc_satisfaction_22'].mean()
mean_dynamic = df_category['dynamic'].mean()
```

Построим график для наглядности:

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
plt.style.use('seaborn-whitegrid')
plt.suptitle('Категории вопросов: удовлетворённость по годам и в динамике', 
             fontweight='bold', y=1.02)

ax1=plt.subplot(1, 3, 1)
sns.barplot(data=df_category, y='category_description', x='perc_satisfaction_21', color='#aab8d0')
plt.axvline(x=mean_2021, color='black', linestyle='--', label='среднее \n за 2021')
plt.legend(loc='lower right') 
plt.title('2021 год', fontweight='bold', y=1.02)
plt.xlabel('% удовлетворённости')
plt.ylabel('')

ax2=plt.subplot(1, 3, 2, sharex=ax1)
sns.barplot(data=df_category, y='category_description', x='perc_satisfaction_22', color='#829cc3')
plt.axvline(x=mean_2022, color='black', linestyle='--', label='среднее \n за 2022')
plt.legend(loc='lower right') 
plt.title('2022 год', fontweight='bold', y=1.02)
plt.xlabel('% удовлетворённости')
plt.ylabel('')
ax2.set(yticklabels=[])  
ax2.tick_params(left=False)

ax3=plt.subplot(1, 3, 3, sharex=ax1)
sns.barplot(data=df_category, y='category_description', x='dynamic', color='#aab8d0')
plt.axvline(x=mean_dynamic, color='black', linestyle='--', label='среднее \n по динамике')
plt.legend(loc='lower right') 
plt.title('динамика', fontweight='bold', y=1.02)
plt.xlabel('% удовлетворённости')
plt.ylabel('')
ax3.set(yticklabels=[])  
ax3.tick_params(left=False)

plt.show();
```

- выше всего родители оценивают доступность объектов, мед-соц экспертизу и общие вопросы: показатели этих категорий более чем в 2 раза превышают среднее значение в 18% удовлетворённости в 2022 году


- средняя удовлетворённость растёт: от 9% в 2021 году к 18% в 2022 году


- чаще всего отрицательно отвечают на вопросы, связанные с: профориентацией, трудом/занятостью, правовой помощью, общественными организациями, паллиативной помощью, проживанием,  санаторно-курортным лечением, образованием (это минимальные показатели 2022 года)


- в основном все показатели растут в среднем на 9% в год, но есть исключение: образование показывает отрицательную динамику

### 10 элементов окружающей среды

Создаём датафрейм:

```python
df_environment_21 = (df_21.groupby('10_elements_of_environment', as_index=False)
                   .agg({'perc_satisfaction_21':'mean', 'question_number':'count'})
                   .rename(columns={'question_number':'question_cnt_21'}))
df_environment_22 = (df_22.groupby('10_elements_of_environment', as_index=False)
                   .agg({'perc_satisfaction_22':'mean', 'question_number':'count'})
                   .rename(columns={'question_number':'question_cnt_22'}))
df_environment = df_environment_22.merge(df_environment_21, on='10_elements_of_environment', how='left')
df_environment['dynamic'] = df_environment['perc_satisfaction_22'] - df_environment['perc_satisfaction_21']
df_environment = df_environment.sort_values(by='perc_satisfaction_22', ascending=False)
```

Снова рассчитаем среднее:
